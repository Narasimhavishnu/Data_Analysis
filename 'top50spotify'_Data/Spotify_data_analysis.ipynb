```python
#problem 2 Write a Python program to perform the following 
#tasks-1.Create a pandasseries from the below dictionary where indices are subjects:
new_series={'English':{'Sam':60,'Jackson':74,'Ahree':85}, 'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}}
n=pd.Series(new_series)
n=n.apply(pd.Series)
n
```

```python
#task 2 convert the series to adtaframes and fill null values to 0
new_df=pd.DataFrame(n)
re_NAN=new_df.fillna(0)
re_NAN
```

```python
#3.Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects
df=re_NAN.transpose()
n=list(df.mean(axis=1))
df['Average']=n
df
```
